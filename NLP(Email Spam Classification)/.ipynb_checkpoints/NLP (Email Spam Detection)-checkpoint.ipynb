<h3><font color=blue face=verdona><center>
Tokenization, Lowercasing, Removing punctuation, special characters<br> and stop words, Stemming

```python
ps = PorterStemmer()

def preprocessing_text(text):
    text = re.sub('[^a-zA-Z]', ' ', text).lower()
    words = text.split()
    words = [ps.stem(word) for word in words if word not in stopwords.words('english') ]
    return ' '.join(words)
```

```python
# preprocessed_data['Main_Content']  = preprocessed_data['Main_Content'].apply(preprocessing_text)
# preprocessed_data.to_csv('processed_data.csv')
preprocessed_data = pd.read_csv('processed_data.csv',index_col=0)
preprocessed_data
```

<h3><font color=blue face=verdona><center>
Tokenization, Lowercasing, Removing punctuation, special characters<br> and stop words, Lemmatizing

```python
lem = WordNetLemmatizer()
def preprocessing_text2(text):
    text = re.sub('[^a-zA-Z]', ' ', text).lower()
    words = text.split()
    words = [lem.lemmatize(word) for word in words if word not in stopwords.words('english') ]
    return ' '.join(words)
```

<center><div style="background-color: brown"><h1><font color=white face="georgia">Train Test Split<br>
    Various Models Training and Evaluation

```python
from sklearn.model_selection import train_test_split, KFold
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import numpy as np
from colorama import Fore
import joblib
from sklearn.metrics import log_loss, accuracy_score,confusion_matrix,precision_score

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier, StackingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from xgboost import XGBRFClassifier,XGBClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA

# scaler = MinMaxScaler((0,1))
# pca = PCA(n_components=20)

gb = GradientBoostingClassifier()
rforest = RandomForestClassifier() 
l_reg = LogisticRegression()
svc = SVC(kernel='linear')
xgb = XGBClassifier()
xgbrf = XGBRFClassifier()
dtree = DecisionTreeClassifier()
knc = KNeighborsClassifier()
gnb = GaussianNB()
mnb = MultinomialNB()

kfold = KFold(n_splits=5, shuffle=True, random_state=42)

models = [gb, rforest, l_reg,svc, xgb, xgbrf, dtree, knc, gnb, mnb]
model_names = [ 'Gradient Boosting', 'RForest', 'Logistic Regression' , 'SVC', 'XGB','XGBRF', 'DTree', 'KNNeighbors', 
               'GNB', 'MNB']
best_indices_dict = {}

for i, model in enumerate(models):
    model_name = model_names[i]
    best_accuracy = 0.0
    print('---------------------------',model_name,'------------------------------')
    mean_val_acc, mean_test_acc = [], []

    for train_indices, test_indices in kfold.split(X):            
            X_train, X_val, y_train, y_val = train_test_split(X[train_indices], y[train_indices], test_size=0.2, random_state=42)

#             X_train = scaler.fit_transform(X_train)
#             X_train = pca.fit_transform(X_train)

#             X_val = scaler.transform(X_val)
#             X_val = pca.transform(X_val)

            model.fit(X_train, y_train)

            y_val_pred = model.predict(X_val)
            
            val_accuracy = accuracy_score(y_val, y_val_pred)
            mean_val_acc.append(val_accuracy)
    
            # Perform testing on the test set for this fold
            X_test, y_test = X[test_indices], y[test_indices]
            
#             X_test = scaler.transform(X_test)
#             X_test = pca.transform(X_test)      
                
            y_test_pred = model.predict(X_test)
            test_accuracy = accuracy_score(y_test, y_test_pred)
            mean_test_acc.append(test_accuracy)

            if val_accuracy > test_accuracy:
                print(f'{Fore.GREEN}Validation accuracy : :', round(val_accuracy, 4), end=f"{Fore.RESET}\t")
                print(f'{Fore.RED}Test accuracy : :', round(test_accuracy, 4), f"{Fore.RESET}")
            else:            
                print(f'{Fore.RED}Validation accuracy : :', round(val_accuracy, 4), end=f"{Fore.RESET}\t")
                print(f'{Fore.GREEN}Test accuracy : :', round(test_accuracy, 4), f"{Fore.RESET}")

            if test_accuracy > best_accuracy:
                best_accuracy = test_accuracy
                best_indices_dict[model_name]={
                    'train': train_indices,
                    'test': test_indices
                }
    print('Mean Val accuracy : ', np.mean(mean_val_acc), 'Mean Test accuracy : ',np.mean(mean_test_acc))
```

<h3><font color=blue face=verdona><center> The SVC model excells all the models with an average of more than<font color=green face=verdona> 99% accuracy</font>, or can say less then even<font color=red face=verdona> 1% error</font>, Making it The best model

<center><div style="background-color: brown"><h1><font color=white face="georgia"> Building Preprocessor Pipeline and Model Pipeline

```python
#access the predictor.py file to view the functions 
#these functions are the steps taken in the project 
from predictor import importer, duplicate_remover, parser, content_extractor, array_transformer
from predictor import preprocessing_text_2
```

```python
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import Pipeline

tf = importer()

preprocessing_pipeline = Pipeline(steps=[('1', FunctionTransformer(parser) ),
                       ('2', FunctionTransformer(content_extractor)),
                       ('3', FunctionTransformer(preprocessing_text_2)),
                       ('4', tf),
                       ('5', FunctionTransformer(array_transformer))])
```

 Applying Preprocessing Pipelines

Building Model in a Pipeline <br> Training Pipeline Model

```python
model_pipeline = Pipeline(steps=[('Stacking Classifier', StackingClassifier(estimators=[('Rf_clf', rforest),
                                     ('Log_Reg', l_reg),
                                      ('svc', svc),
                                      ('XGB_clf', xgb),
                                     ('MNB', mnb)], final_estimator=svc))])

model_pipeline.fit(x_train, y_train)
```
